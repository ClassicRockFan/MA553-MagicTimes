```python
#Use a tensorflow estimator from sagemaker to train model
import sagemaker

import tensorflow as tf
tf_version = tf.__version__

from sagemaker.tensorflow import TensorFlow

#Sagemaker role, make sure you've allowed access to any S3 bucket
role = sagemaker.get_execution_role()
role


tf_estimator = TensorFlow(entry_point='train.py', 
                          role=role,
                          instance_count=1, 
                          instance_type='ml.c5.18xlarge',linear
                          framework_version=tf_version, 
                          py_version='py3',
                          script_mode=True,
                          hyperparameters={
                              'epochs': 15
                          }
                         )

#Training
tf_estimator.fit({'train': "s3://sagemaker-park-data/universal-data/universal_train.pkl"})
```
